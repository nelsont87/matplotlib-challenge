```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_df = pd.read_csv(mouse_drug_data_to_load)
clinical_df = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
#print(mouse_df.head())
#print(clinical_df.head())
merge_table = pd.merge(mouse_df, clinical_df, on = "Mouse ID", how="outer")
merge_table_drop = merge_table.drop(columns=["Metastatic Sites", "Mouse ID"])
merge_table.head()

# Display the data table for preview

```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
groupby_df = merge_table_drop.groupby(["Drug", "Timepoint"], as_index = False)
group_df = groupby_df["Tumor Volume (mm3)"].mean()
group_df.head()


# Convert to DataFrame


# Preview DataFrame
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
grp_sem = merge_table_drop.groupby(["Drug", "Timepoint"]).sem()
grp_sem = grp_sem.reset_index()
grp_sem.head()

# Convert to DataFrame
pivot_df = grp_sem.pivot(index = "Timepoint", columns = "Drug", values = "Tumor Volume (mm3)")

# Preview DataFrame
pivot_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
format_df = group_df.pivot(index = "Timepoint", columns = "Drug", values = "Tumor Volume (mm3)")
# Preview that Reformatting worked
format_df
```

```python
# Generate the Plot (with Error Bars)
ax = format_df["Capomulin"].plot(marker = "o")
ax = format_df["Infubinol"].plot(marker = '^')
ax = format_df["Ketapril"].plot(marker = 's')
ax = format_df["Placebo"].plot(marker = 'd')
errorbar_plot = ax.errorbar(format_df.index, format_df["Capomulin"], yerr = pivot_df["Capomulin"] )
errorbar_plot = ax.errorbar(format_df.index, format_df["Infubinol"], yerr = pivot_df["Infubinol"])
errorbar_plot = ax.errorbar(format_df.index, format_df["Ketapril"], yerr = pivot_df["Ketapril"])
errorbar_plot = ax.errorbar(format_df.index, format_df["Placebo"], yerr = pivot_df["Placebo"])
ylab = ax.set_ylabel('Tumor Volume (mm3)')
title = ax.set_title("Tumor Response to Treatment")
legend = ax.legend()
plt.grid(b=True, which='major', color='#666666', linestyle='-')

# Save the Figure

```

```python
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
group_met = merge_table.groupby(["Drug", "Timepoint"])
group_met_df = group_met["Metastatic Sites"].mean()


# Convert to DataFrame
group_met_df = group_met_df.to_frame()
# Preview DataFrame
group_met_df.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
sem_met = merge_table.groupby(["Drug", "Timepoint"]).sem().drop(columns=["Tumor Volume (mm3)", "Mouse ID"])
# Convert to DataFrame
sem_met.head()
# Preview DataFrame
sem_met_reset = sem_met.reset_index()
sem_met_reset.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
format_met_df = sem_met_reset.pivot(index = "Timepoint", columns = "Drug", values = "Metastatic Sites")

# Preview that Reformatting worked
format_met_df 
```

```python

met_pivot = pd.pivot_table(merge_table, index = "Timepoint", columns = "Drug", values = "Metastatic Sites")
met_pivot
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(met_pivot.index, met_pivot["Capomulin"], yerr = format_met_df["Capomulin"], marker = "o", label = "Capomulin")
plt.errorbar(met_pivot.index, met_pivot["Infubinol"], yerr = format_met_df["Infubinol"], marker = "^", label = "Infubinol")
plt.errorbar(met_pivot.index, met_pivot["Ketapril"], yerr = format_met_df["Ketapril"], marker = "s", label = "Ketapril")
plt.errorbar(met_pivot.index, met_pivot["Placebo"], yerr = format_met_df["Placebo"], marker = "d", label = "Placebo")
plt.legend()
# Save the Figure
plt.grid(b=True, which='major', color='#666666', linestyle='-')
# Show the Figure
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
count_df = merge_table.groupby(["Drug", "Timepoint"])
count_df_count = count_df["Mouse ID"].count()
count_mouse_df["Mouse Count"] = count_df_count.to_frame()
count_mouse_df_drop = count_mouse_df.drop(columns = ["Mouse ID"])
count_mouse_df_reset = count_mouse_df_drop.reset_index()
count_mouse_df_reset.head()
# Convert to DataFrame

# Preview DataFrame
```

```python
# Minor Data Munging to Re-Format the Data Frames
count_pivot = pd.pivot_table(count_mouse_df_reset, index = "Timepoint", columns = "Drug", values = "Mouse Count")
count_pivot
# Preview the Data Frame
```

```python
# Generate the Plot (Accounting for percentages)
plt.plot(((count_pivot["Capomulin"]/count_pivot["Capomulin"].sum())*100), marker= 'o', label = "Capomulin")
plt.plot(((count_pivot["Infubinol"]/count_pivot["Infubinol"].sum())*100), marker = '^', label = "Infubinol")
plt.plot(((count_pivot["Ketapril"]/count_pivot["Ketapril"].sum())*100), marker = 's', label = "Ketapril")
plt.plot(((count_pivot["Placebo"]/count_pivot["Placebo"].sum())*100), marker = 'd', label = "Placebo")
# Save the Figure
plt.grid(b=True, which='major', color='#666666', linestyle='-')
plt.legend()
# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
tumor = pd.pivot_table(merge_table, index = "Timepoint", columns = "Drug", values = "Tumor Volume (mm3)")
percent_change = tumor.pct_change(periods = 9) * 100
# Display the data to confirm
change = percent_change.iloc[-1] 
change
```

```python
# Store all Relevant Percent Changes into a Tuple
tup = percent_change.iloc[-1, [0,2,3,5]]
tupx = tuple(tup)
tupx
x_axis = np.arange(len(tupx))
# Splice the data between passing and failing drugs
plt.bar(x_axis, tupx, color=(tup > 0).map({True: 'r',False: 'g'}), width = -1, align = "edge")

# Orient widths. Add labels, tick marks, etc. 

tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, ["Capomulin", "Infubinol", "Ketapril", "Placebo"])
# Use functions to label the percentages of changes


# Call functions to implement the function calls


# Save the Figure

plt.title("Tumor Change Over 45 Day Treatment")
plt.ylabel("% Tumor Volume Change")
plt.grid(b=True, which='major', color='#666666', linestyle='-')
plt.ylim(-25, 63)
# Show the Figure
#fig.show()
```
